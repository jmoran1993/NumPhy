```python
%pylab inline 
import math, random
import numpy as np 
```

```python
Ntrials = 10000 
random.seed()
alpha = 3. 
xmeanexact = alpha /(alpha-1.)
xicontrol = []
xtot = 0. 
for itime in range(Ntrials):
    xi = random.random()**(-1/alpha)
    xtot += xi
    xicontrol.append( xtot / float(itime + 1))
pylab.title('Quality Control')
pylab.xlabel('time')
pylab.ylabel('running average of xi')
time = [t for t in range(Ntrials)]
meanexact = [ xmeanexact for t in range(Ntrials)]
pylab.plot(xicontrol, 'r-')
pylab.plot(time, meanexact, 'k-')
pylab.axis([0, Ntrials, xmeanexact * 0.9, xmeanexact * 1.1])
pylab.show()
```
